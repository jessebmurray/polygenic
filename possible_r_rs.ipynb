```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_iter = 10_000
r = np.linspace(0, 1, num_iter)
```

```python
rs_lb = np.sqrt(r*(1-r))
```

```python
rs_sp = np.sqrt(1-r**2)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

plt.plot(r, rs_lb, linestyle='--', color='black', alpha=0.8,
        label='Degenerate\nlower bound')
plt.plot(r, rs_sp, label='Stable population\nvariance')

plt.fill_between(r, rs_lb, 5*np.ones(num_iter),
                alpha=0.4, color='grey')

ax.set(xlim=(0, 1), ylim=(0, 1.2))
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel(xlabel='$r$', fontsize=12)
plt.ylabel(ylabel='$r_s$', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
# plt.savefig('latex/figures/possible_r_rs.png', dpi=300)
plt.show()
```
