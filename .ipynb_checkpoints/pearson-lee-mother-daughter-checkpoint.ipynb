### Load data

```python
plots_path = 'latex/figures/'
data_path = 'data/'
# Round number
rn = 5  
```

Karl Pearson organized the collection of data on over 1100 families in England in the period 1893 to 1898. This particular data set gives the heights in inches of parents and their children (of the same sex), with up to two children per parent. All children are at least age 18, and all parents are younger than 65. Data were given in the source as a frequency table to the nearest inch. Rounding error has been added to remove discreteness from graph.

```python
df = pd.read_csv(data_path + file_name + '.csv')

# The raw (unstandardized) parent data
x_raw = df.Parent.to_numpy()

# The raw (unstandardized) child data
y_raw = df.Child.to_numpy()
```

```python
df.head()
```

```python
# Get the standard deviation of the parents in cm 
x_raw.std(ddof=1) * 2.54
```

```python
# Get the standard deviation of the children in cm 
y_raw.std(ddof=1) * 2.54
```

```python
# There are the same number of parent-child paris
n = x_raw.size
assert n == x_raw.size == y_raw.size

# The number of parent-child pairs
n
```

### Standardize to the parent distribution

```python
# Get the mean and SD of the parent distribution
x_sample_sd = np.std(x_raw, ddof=0)
x_sample_mean = np.mean(x_raw)
```

```python
# Standardize the parent and children and scores to this
# mean and variance
x = (x_raw - x_sample_mean) / x_sample_sd
y = (y_raw - x_sample_mean) / x_sample_sd
```

```python
# Test that x is scaled correctly (mean 0, variance 1)
assert np.allclose(x.mean(), 0)
assert x.var() == 1
```

```python
mv.scatter_plot(x, y)
```

### Estimate $r$ and $r_s$

Calcualte the residuals from the normal linear model equation
$$y = r x + c+ \epsilon$$

$$\epsilon \sim \mathcal{N}(0, r_s^2 \sigma^2)$$

$$\sigma^2 = 1$$

$$\hat{r}_s = \sqrt{RSS / (n-p)}$$
$$p = 2$$

```python
from sklearn.linear_model import LinearRegression

lmb = LinearRegression(fit_intercept=True)
lmb.fit(x.reshape(-1, 1), y)
```

```python
# Calculate the regression coefficient r
r = lmb.coef_[0]

# Calculate the intercept
c = lmb.intercept_
```

```python
# Test the intercept is the same as the y mean
assert np.allclose(c, np.mean(y))
```

```python
# Manually calculate the beta coefficients
X = np.concatenate([[np.ones(x.size)], [x]]).T
Y = y.reshape(-1, 1)

beta = np.linalg.inv(X.T @ X) @ X.T @ Y
beta = beta.ravel()

# Test that the number of coefficients is two 
# (the intercept  and the coefficient)
p = 2
assert beta.size == p

# Test that the manual calculation agrees with 
# the SkLearn software calculation
assert np.allclose(beta, (c, r))
```

```python
# Get the residuals (see the equation)
resid = y - r*x - c
assert resid.size == n
```

```python
# Calculate the residual coefficient estimate (see the equation)
# The numerator is the sum of squares (dot product)
rs = np.sqrt(np.dot(resid, resid) / (n - p))

# Check that it's close to the standard deviaiton of the residuals
assert np.allclose(resid.std(), rs, atol=0.01)
```

```python
print('Intercept c (in cm):', np.round(c * x_sample_sd * 2.54, rn))
print()
print('Regression coefficient r:', np.round(r, rn))
print()
print('Residual coefficient r_s:', np.round(rs, rn))
```

### Check if the residuals vary with the parent score

```python
plt.figure(figsize=(6, 6))
plt.scatter(x, resid, alpha=0.5,
           facecolors='none', edgecolors='#1f77b4')
plt.axhline(y=0, color='grey')
plt.grid(alpha=0.4)
plt.xlabel('Parent score')
plt.ylabel('Child residual')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
bins = np.arange(-2, 2.5, 0.5)
x_binned = np.digitize(x, bins)

resid_means = np.array([resid[x_binned == i].mean() - resid.mean() for i in range(bins.size+1)])
resid_sds = np.array([resid[x_binned == i].std() for i in range(bins.size+1)])

true_bins = np.arange(-2.25, 2.75, 0.5)
```

#### SD confidence interval

```python
plt.figure(figsize=(6, 6))
plt.plot(true_bins, resid_means, color='black')
plt.errorbar(true_bins, resid_means,
            yerr=resid_sds, color='grey', alpha=0.8)
plt.axhline(y=0, color='grey')
plt.xlabel('Parent score')
plt.ylabel('Child residual')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

#### Normal confidence interval

```python
# Get the confidence intervals
resid_cis = np.array([st.norm.interval(alpha=0.95, 
                        loc=np.mean(resid[x_binned == i] - resid.mean()), 
                    scale=st.sem(resid[x_binned == i] - resid.mean())) 
                     for i in range(bins.size+1)])

resid_cis = np.abs(resid_cis - resid_means.reshape(-1, 1))
```

```python
mv.plot_residuals_by_parent(true_bins, resid_means, resid_cis)
if save_plots:
    plt.savefig(plots_path + file_name + '-residuals_by_parent.png', dpi=300)
plt.show()
```
