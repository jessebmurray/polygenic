```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statistics
```

```python
plt.style.use('seaborn')
plt.style.use('seaborn-talk')
```

```python
def polygenic_pop(num_genes=700, population=3000, effect='distributed'):
    phenotypes = list()
    for population_member in range(population):
        total_effect = 0
        for gene_num in range(num_genes):
            if effect == 'absolute':
                gene_effect = np.random.choice([-1, 1])
            elif effect == 'distributed':
                gene_effect = np.random.randn()
            elif effect == 'uniform':
                gene_effect = np.random.uniform(low=-2, high=2)
            total_effect += gene_effect

        phenotypes.append(total_effect)
    # plt.hist(effects)
    print('Standard deviation:', statistics.stdev(phenotypes))
    sns.distplot(phenotypes)
    plt.show()
    
    return phenotypes
```

```python
absolute = polygenic_pop(num_genes=700, population=1000, effect='absolute')
```

```python
distributed = polygenic_pop(num_genes=700, population=1000, effect='distributed')
```

```python
uniform = polygenic_pop(num_genes=700, population=1000, effect='uniform')
```

They're all roughly normal because of the central limit theorem. As a result, when there are many genes, the distribution of the individual effects of the genes doesn't matter.
